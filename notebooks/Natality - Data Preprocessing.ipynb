###### Feature Engineering

---------

    - New Column for Birth Year and Month

```python
df0['birth_year_month'] = df0['birth_year'].astype(int).astype(str) + df0['birth_month'].astype(int).astype(str).str.zfill(2)
df0['birth_year_month'] = pd.to_datetime(df0['birth_year_month'], format = '%Y%m')
```

    - Creating/Binning groups for Mother's BMI

```python
df0['mother_bmi_recode'] = pd.cut(df0['mothers_bmi'],[0.0,19.0,25.0,30.0,35.0,40.0,100.0])
```

```python
df0['mothers_age_groups']= pd.cut(df0['mothers_age'], [0,14,19,24,29,34,39,44,49,100],
                                     labels = ['<15', '15-19', '20-24','25-29','30-34','35-39','40-44','45-49','50+'])
```
